# Hyperparameter Tuning
This notebook interprets the final modelâ€™s predictions using SHAP values and feature importance techniques to understand what drives Top 20 finishes in the Tour de France.

## Key Steps:
- Uses SHAP to visualize global and individual feature impact
- Analyses feature importance to explain prediction drivers

## Import Libraries

```python
import pandas as pd
import numpy as np
import sys
from pathlib import Path

# Scikit-learn
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import StratifiedKFold, GridSearchCV
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.metrics import (
    classification_report, confusion_matrix, roc_auc_score, roc_curve,
    precision_recall_curve
)

# Plotting
import matplotlib.pyplot as plt

import shap
shap.initjs()
```

## Set Folder Path and Read CSVs

```python
def find_project_root(start: Path, anchor_dirs=("src", "Data")) -> Path:
    """
    Walk up the directory tree until we find a folder that
    contains all anchor_dirs (e.g. 'src' and 'Data').
    """
    path = start.resolve()
    for parent in [path] + list(path.parents):
        if all((parent / d).is_dir() for d in anchor_dirs):
            return parent
    raise FileNotFoundError("Could not locate project root")
```

```python
# Locate the project root regardless of notebook depth
project_root = find_project_root(Path.cwd())

# ----- Code modules --------------------------------------------------
src_path = project_root / "src"
if str(src_path) not in sys.path:
    sys.path.append(str(src_path))

from data_prep import preprocess_tdf_data   # import data preproc function

# ----- Data ----------------------------------------------------------
raw_data_path = project_root / "Data" / "Raw"
processed_data_path = project_root / "Data" / "Processed"
print("Raw data folder:", raw_data_path)
print("Processed data folder:", processed_data_path)
```

```python
# Race metadata
prepared_df = pd.read_csv(processed_data_path / "tdf_prepared_2011_2024.csv")
```

## Clean Data

```python
# import missing_value_handler
from missing_value_handler import FillWithSentinel
```

```python
cleaner = FillWithSentinel()
final_df = cleaner.fit_transform(prepared_df)
```

```python
# Filter out DNF or DSQ from TDF_Pos
final_df = final_df[~final_df['TDF_Pos'].isin(['DNF', 'DSQ'])]
```

```python
final_df = final_df.dropna(subset=['TDF_Pos'])
```

```python
# Convert TDF_Pos to numeric
final_df['TDF_Pos'] = pd.to_numeric(final_df['TDF_Pos'])

# 1 if TDF_Pos <= 20, else 0
final_df['is_top20'] = (final_df['TDF_Pos'] <= 20).astype(int)
```

```python
# Set date range for 2015+, and exclude 2020 & 2021
final_df = final_df[(final_df['Year'] >= 2015) & (~final_df['Year'].isin([2020, 2021]))]
```

```python
final_df
```

## Test Model

```python
features = ['Best_Pos_BT_UWT', 'Best_Pos_BT_PT',
       'FC_Pos_YB', 'best_recent_tdf_result',
       'best_recent_other_gt_result', 'rode_giro']
```

```python
# --------------------------------------------
# 0.  Prep â€“ train/test masks + splitter
# --------------------------------------------
train_mask = (final_df['Year'] <= 2023)
test_mask  = (final_df['Year'] == 2024)

cv_splitter = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
X_train = final_df.loc[train_mask, features]
y_train = final_df.loc[train_mask, 'is_top20']
X_test  = final_df.loc[test_mask,  features]
y_test  = final_df.loc[test_mask,  'is_top20']
```

```python
pipeline = Pipeline([
    ('imputer', SimpleImputer()),
    ('classifier', GradientBoostingClassifier(random_state=42))
])

param_grid = {
    'classifier__n_estimators': [100, 200],
    'classifier__learning_rate': [0.01, 0.05],
    'classifier__max_depth': [2, 3],
    'classifier__min_samples_split': [2, 5],
    'classifier__min_samples_leaf': [1, 2],
    'classifier__subsample': [0.8],
    'classifier__max_features': ['sqrt']
}


gs = GridSearchCV(
    pipeline,
    param_grid,
    scoring='roc_auc',
    cv=5,
    n_jobs=-1,
    verbose=2
)

gs.fit(X_train, y_train)
```

## Analyse Results

```python
# Predict
y_pred = gs.best_estimator_.predict(X_test)
y_proba = gs.best_estimator_.predict_proba(X_test)[:, 1]

# Confusion matrix
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Classification report
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

# AUC
test_auc = roc_auc_score(y_test, y_proba)
print(f"\nTest AUC: {test_auc:.4f}")

# ROC Curve
fpr, tpr, _ = roc_curve(y_test, y_proba)
plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, label=f'ROC Curve (AUC = {test_auc:.2f})')
plt.plot([0, 1], [0, 1], '--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Precision-Recall Curve
prec, rec, _ = precision_recall_curve(y_test, y_proba)
plt.figure(figsize=(6, 4))
plt.plot(rec, prec, label="Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
model = gs.best_estimator_.named_steps['classifier']
importances = model.feature_importances_
feat_names = X_train.columns

importance_df = pd.DataFrame({'Feature': feat_names, 'Importance': importances})
importance_df.sort_values('Importance', ascending=True).plot.barh(x='Feature', figsize=(8, 5))
plt.title("Feature Importances")
plt.tight_layout()
plt.show()
```

```python
# Get predictions on test set
y_pred = gs.best_estimator_.predict(X_test)
y_proba = gs.best_estimator_.predict_proba(X_test)[:, 1]  # Probability of class 1 (top 20)

# Create a dataframe to merge with final_df
pred_df = pd.DataFrame({
    'Predicted_Label': y_pred,
    'Predicted_Probability': y_proba
}, index=X_test.index)  # Same index as X_test so we can join directly

# Join back to the original final_df
final_df_with_preds = final_df.join(pred_df)
```

```python
final_df_with_preds = final_df_with_preds.loc[test_mask, ['Rider_ID', 'TDF_Pos'] + ['Predicted_Probability'] + features]
```

## SHAP Analysis

```python
# Use TreeExplainer for tree-based models (e.g. XGBoost, LightGBM)
explainer = shap.Explainer(model)  # or shap.TreeExplainer(model) if preferred
```

```python
# Select rider IDs to explain
rider_ids = [27332, 27333, 37421, 81835, 26504]
```

```python
# Extract features for SHAP
riders_to_explain = final_df[(final_df['Rider_ID'].isin(rider_ids)) & (final_df['Year']==2024)][features]
shap_values = explainer(riders_to_explain)
```

```python
# Pull info from final_df_with_preds
info_df = final_df_with_preds[final_df_with_preds['Rider_ID'].isin(rider_ids)][['Rider_ID', 'TDF_Pos', 'Predicted_Probability']]

# Make sure the info_df is in the same order as rider_ids
#info_df = info_df.set_index('Rider_ID').loc[rider_ids].reset_index()
```

```python
for i in range(len(rider_ids)):
    rider_info = info_df.iloc[i]
    rider_id = int(rider_info.iloc[0])
    rider_pos = int(rider_info.iloc[1])
    rider_prob = rider_info.iloc[2]

    print("======== Rider:", rider_id, ", Probability:", rider_prob, ", Actual Pos:", rider_pos, "========")
    # Get row from full DataFrame
    row = final_df[final_df['Rider_ID'] == rider_id]
    
    # Get predictions and values
    rider_values = final_df_with_preds[final_df_with_preds["Rider_ID"] == rider_id]
    
    # Display info
    display(rider_values)
    
    # Calculate SHAP values for the current rider
    shap_val = explainer(row[features])
    
    # Build a DataFrame of SHAP values
    shap_df = pd.DataFrame({
        'Feature': features,
        'SHAP Value': shap_val.values[0],
        'Feature Value': row[features].values[0]
    }).sort_values('SHAP Value', ascending=False)
    
    display(shap_df)
    
    # Show force plot
    display(shap.plots.force(shap_val[0]))
```
